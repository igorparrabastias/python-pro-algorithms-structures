# Cálculo del Tamaño de un Árbol Binario

El tamaño de un árbol binario se refiere a la cantidad total de nodos presentes en el árbol. Este valor es fundamental para entender la escala de un árbol y es relevante en la evaluación de la eficiencia de varias operaciones y algoritmos que involucran árboles binarios.

- **Concepto de Tamaño de un Árbol Binario:**
  - El tamaño de un árbol se puede definir como el número total de sus nodos.
  - Se puede calcular mediante un recorrido por el árbol, contando cada nodo que se visita.

- **Aplicaciones del Tamaño de un Árbol:**
  - Evaluar la complejidad de algoritmos que dependen del número total de nodos.
  - Ayudar en la planificación de operaciones que involucran la estructura completa del árbol, como el balanceo o la serialización.
  - Utilizar en el diagnóstico del estado y forma del árbol, por ejemplo, para determinar si un árbol está lleno o completo.

## Implementación en Python

Aquí te muestro cómo implementar este método en la clase [`BinaryTree`](src/BinaryTree.py): 

```python
# Código utilitario
from src.visualization import visualize_bt
# ver src/BinaryTree.py
from src.BinaryTree import BinaryTree
```

```python
def size(self, node):
    # Si el nodo actual es None, no cuenta para el tamaño.
    if node is None:
        return 0
    # Cuenta el nodo actual más el tamaño de los subárboles izquierdo y derecho.
    return 1 + self.size(node.left) + self.size(node.right)

# Extender la clase BinaryTree con el nuevo método size
BinaryTree.size = size
```

## Pruebas de `size`

Verifiquemos la funcionalidad del método `size` creando un árbol y calculando su tamaño.

```python
# Crear un árbol y agregar nodos
tree = BinaryTree()
tree.insert(1)
tree.insert(2)
tree.insert(3)
tree.insert(4)
tree.insert(5)

# Calcular el tamaño del árbol
tamanio = tree.size(tree.root)
print(f"El tamaño del árbol es: {tamanio}")

# Visualizar el árbol
dot = visualize_bt(tree)
dot
```

## Complejidad del Algoritmo

- **Complejidad Temporal:** El cálculo del tamaño tiene una complejidad de tiempo O(N), donde N es el número total de nodos en el árbol, ya que cada nodo necesita ser visitado exactamente una vez.

- **Complejidad Espacial:** La complejidad espacial es O(h), donde h representa la altura máxima del árbol. Esto se debe al almacenamiento de las llamadas recursivas en la pila de ejecución. Para un árbol completamente balanceado, esto sería O(log N), pero en el peor caso (un árbol degenerado), sería O(N).

## Ejercicios Prácticos

1. Implemente una función que calcule la diferencia de tamaño entre el subárbol izquierdo y el subárbol derecho para cada nodo y que devuelva la suma de todas estas diferencias.
2. Escriba una función que determine si el tamaño de un árbol binario es un número par o impar.

## Soluciones a los Ejercicios

```python
# **Ejercicio 1:**

def size_difference(node):
    if not node:
        return 0, 0  # No difference and no size

    left_diff, left_size = size_difference(node.left)
    right_diff, right_size = size_difference(node.right)

    total_diff = left_diff + right_diff + abs(left_size - right_size)
    total_size = 1 + left_size + right_size

    return total_diff, total_size

def total_size_difference(tree):
    total_diff, _ = size_difference(tree.root)
    return total_diff

# **Ejercicio 2:**

def is_size_even(tree):
    return tree.size(tree.root) % 2 == 0
```
