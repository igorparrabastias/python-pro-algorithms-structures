# Anexo 1: Clasificación y Tipos de Algoritmos

- Exploración de los diferentes tipos de algoritmos y su clasificación.

### **Introducción**

Los algoritmos son procedimientos o fórmulas para resolver problemas. Dada la diversidad de problemas que pueden ser abordados mediante algoritmos, existe una amplia variedad de tipos, cada uno adaptado a distintas necesidades y escenarios. La clasificación de los algoritmos puede basarse en diferentes criterios, como su enfoque de solución, complejidad, o el tipo de problema que resuelven. Esta exploración proporciona una visión general de los principales tipos de algoritmos y su clasificación.

### **Clasificación Basada en Enfoque de Solución**

### **Algoritmos de Dividir y Conquistar**

Estos algoritmos resuelven un problema dividiéndolo en subproblemas más pequeños del mismo tipo, resolviendo los subproblemas de manera recursiva y combinando sus soluciones para resolver el problema original. Ejemplos: QuickSort, MergeSort.

### **Algoritmos Greedy (Ávidos)**

Los algoritmos greedy toman decisiones secuenciales, en cada paso eligiendo la opción más óptima sin considerar el problema global, con el objetivo de encontrar una solución general óptima. Ejemplos: Algoritmo de Kruskal, Algoritmo de Dijkstra.

### **Algoritmos de Programación Dinámica**

La programación dinámica es utilizada para resolver problemas de optimización dividiéndolos en subproblemas más simples y almacenando las soluciones de estos subproblemas para evitar el cálculo repetido. Ejemplos: Algoritmo de Fibonacci, Problema de la Mochila.

### **Algoritmos de Backtracking**

Estos algoritmos prueban todas las posibilidades hasta encontrar la solución correcta, retrocediendo paso a paso para corregir errores y probar alternativas. Ejemplo: Problema de las ocho reinas.

### **Clasificación Basada en Tipo de Operación**

### **Algoritmos de Búsqueda**

Son utilizados para encontrar un elemento específico dentro de una estructura de datos. Ejemplos: Búsqueda lineal, Búsqueda binaria.

### **Algoritmos de Ordenación**

Ordenan los elementos de una colección de datos según un orden específico. Ejemplos: Bubble Sort, QuickSort, MergeSort.

### **Algoritmos Numéricos**

Resuelven problemas matemáticos y numéricos. Ejemplos: Algoritmo de Euclides para el MCD, Transformada Rápida de Fourier (FFT).

### **Clasificación Basada en Complejidad**

### **Algoritmos P**

Problemas que pueden ser resueltos en tiempo polinomial por una máquina de Turing determinista. Ejemplo: Algoritmo de Dijkstra.

### **Algoritmos NP**

Problemas cuya solución puede ser verificada en tiempo polinomial, aunque no necesariamente pueden ser resueltos en tiempo polinomial. Ejemplo: Problema de la mochila.

### **Algoritmos NP-Completo**

Un subconjunto de NP para el cual se cree que no existe un algoritmo que los resuelva en tiempo polinomial. Ejemplo: Problema del viajante.

### **Conclusión**

La clasificación de algoritmos ofrece una manera de entender y comparar diferentes enfoques para resolver problemas. Al elegir un algoritmo, es importante considerar la naturaleza del problema, los recursos disponibles y los requisitos de eficiencia. La comprensión de los distintos tipos de algoritmos y su clasificación es fundamental para cualquier profesional en el campo de la ciencia de la computación y la ingeniería de software, permitiendo el desarrollo de soluciones eficaces y optimizadas.
